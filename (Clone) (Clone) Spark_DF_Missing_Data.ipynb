```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("miss").getOrCreate()

df = spark.read.csv('/Volumes/udemy/default/sales/ContainsNull/', inferSchema = True, header = True)
df.show()
```

### Drop Missing Data

```python
df.na.drop().show()

```

```python
# Do not drop null values
#df.na.drop(how = 'none').show() 

# Drop row if all the values has null
df.na.drop(how = 'all').show()

# Only drop when any columns is null
df.na.drop(how = 'any').show()
```

```python
# Do not drop if two values are null
df.na.drop(how = 'any', thresh = 2).show()
```

```python

df.na.drop(how = 'any', subset = ['Sales']).show()
df.na.drop(how = 'all', subset = ['Sales']).show()
```

### Set Default values 

```python
df.printSchema()

# Set default value in String column 
df.na.fill('No Name').show()
```

```python
df.na.fill(0).show()
df.na.fill('No Name', 'Name').show()
#or
df.na.fill('No Name', subset = 'Name').show()
```

```python
from pyspark.sql.functions import mean

mean_value = df.select(mean('Sales')).collect()
mean_value 
```

```python
mean_value[0][0]
```

### Replace Null values with "Mean vlue"

```python
mean_value = mean_value[0][0]

df.na.fill(mean_value,['Sales']).show()
```

```python
# will give the same result
df.na.fill(df.select(mean('Sales')).collect()[0][0],['Sales']).show()

# Find breakup below:
```
